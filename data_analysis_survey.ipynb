## 4. Years of Coding Experience
Is there a way to see how many years of coding experience the respondents have? And how many years they've been coding professionally? Can we ignore the people who didn't answer these questions? And can we compare these two groups?

```python
# Inspecting unique values for 'YearsCode'
df['YearsCode'].unique()
```

```python
# Inspecting unique values for 'YearsCodePro'
df['YearsCodePro'].unique()
```


In this section, we are going to analyze the coding experience of the survey respondents. We are interested in two aspects:

- The overall years of coding experience.
- The years of professional coding experience.

To do this, we first define a function `convert_years` that converts the years of experience into a numerical format. This function handles special cases` like 'Less than 1` year`' and 'More than 50 `years'.

We then apply this function` to the 'Ye`arsCo`de' and 'Years`CodePro' columns of our dataframe to convert them into numerical format. We d`rop `any 'NA' values during this process.

Next, we create a function `plot_histogram` that takes in the data to be plotted, the subplot axis (`ax`), the title of the plot, and the color of the histogram. This function is used to plot the histograms for `years_code` and `years_code_pro`, eliminating the duplicated code and making the code cleaner and more mainhat we find!


```python
def convert_years(x):
    if x == 'Less than 1 year':
        return 0
    elif x == 'More than 50 years':
        return 51
    else:
        return pd.to_numeric(x, errors='coerce')

#52 unique values
def plot_histogram(data, ax, title, color):
    ax.hist(data, bins=52, color=color, edgecolor='black')
    ax.set_ylabel('Number of Respondents')
    ax.set_xlabel('Years of Coding')
    ax.set_title(title)

years_code = df['YearsCode'].apply(convert_years).dropna()
years_code_pro = df['YearsCodePro'].apply(convert_years).dropna()

fig, axs = plt.subplots(1, 2, figsize=(20,6))

plot_histogram(years_code, axs[0], 'Distribution of Respondents by Years of Coding', 'blue')
plot_histogram(years_code_pro, axs[1], 'Distribution of Respondents by Years of Pro Coding', 'red')

plt.show()
```

The histograms provide us with valuable insights into the coding experience of the respondents:

- The first histogram, which represents the overall years of coding experience, shows that a significant number of respondents have less than 10 years of experience. This suggests that a large portion of the respondents are relatively new to coding.

- The second histogram, which represents the years of professional coding experience, shows a similar trend. However, the number of respondents with less than 10 years of professional coding experience is slightly less than the overall coding experience. This could be because some respondents might have started coding as a hobby or as part of their education before transitioning into a professional role.

- In both histograms, we see a decline in the number of respondents as the years of experience increase. This indicates that there are fewer individuals with extensive experience in coding.

These findings are important as they give us an understanding of the experience level of the respondents. This information can be useful for organizations and educators in tailoring their programs and resources to meet the needs of their audience.

## 7. Compensation and Coding Experience
Is there a connection between how much a respondent earns and how much coding experience they have?

```python
# Define a function to calculate IQR
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
```

In this section, we are going to analyze the correlation between the years of coding experience and the yearly compensation of the respondents. We are particularly interested in four currencies: USD (United States Dollar), EUR (European Euro), INR (Indian Rupee), and GBP (Pound Sterling).

To do this, we first filter our dataframe to include only the respondents who reported their compensation in one of these four currencies. We also select only the `'Currency'`, `'YearsCode'`, and `'ConvertedCompYearly'` columns for our analysis.

Next, we clean our data by replacing the `'Less than 1 year'` and `'More than 50 years'` values in the `'YearsCode'` column with 0 and 51, respectively. We also convert the 'YearsCode' and 'ConvertedCompYearly' columns to integer data types and drop any 'NA' values.

We then remove any outliers in our data using the `remove_outliers` function.

Finally, we group our cleaned and filtered data by 'Currency' and calculate the correlation between `'YearsCode'` and `'ConvertedCompYearly'` within each group. The result is a dataframe that shows the correlation between years of coding experience and yearly compensation for each of the four currencies.

```python
# Define the currencies of interest
currencies = ["USD United States dollar", "EUR European Euro", "INR Indian rupee", "GBP Pound sterling"]

# Filter the DataFrame by currencies and select only the specified columns
new_df_YearsCode = df[df['Currency'].isin(currencies)][['Currency', 'YearsCode', 'ConvertedCompYearly']].copy()

# Cleaning
new_df_YearsCode = new_df_YearsCode.dropna()
new_df_YearsCode['YearsCode'] = new_df_YearsCode['YearsCode'].replace({'Less than 1 year' : 0, 'More than 50 years' : 51})
new_df_YearsCode['YearsCode'] = new_df_YearsCode['YearsCode'].astype(int)
new_df_YearsCode['ConvertedCompYearly'] = new_df_YearsCode['ConvertedCompYearly'].astype(int)

# Removing outliers
new_df_no_out = remove_outliers(new_df_YearsCode, 'YearsCode')
new_df_no_out = remove_outliers(new_df_YearsCode, 'ConvertedCompYearly')

# Group the DataFrame by 'Currency' and calculate the correlation within each group
correlation = new_df_no_out.groupby('Currency').apply(lambda x: x[['YearsCode', 'ConvertedCompYearly']].corr())
correlation
```

In this section, we are visualizing the correlation between coding experience and compensation for each of the four currencies: USD, EUR, INR, and GBP. We create a scatter plot for each currency, showing the relationship between years of coding experience (x-axis) and yearly compensation (y-axis). The title of each plot includes the currency and the correlation value, giving us a quick overview of the strength and direction of the relationship.


```python
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
axes = axes.ravel()

for i, currency in enumerate(currencies):
    subset = new_df_no_out[new_df_no_out['Currency'] == currency]
    corr_value = correlation.loc[currency, 'YearsCode']['ConvertedCompYearly']
    sns.scatterplot(data=subset, x='YearsCode', y='ConvertedCompYearly', ax=axes[i])
    axes[i].set_title(f'{currency} (correlation: {corr_value:.2f})')

plt.tight_layout()
plt.show()
```


The scatter plots provide us with valuable insights into the correlation between years of coding experience and yearly compensation for each of the four currencies:

- For the **USD (United States Dollar)**, there is a weak positive correlation of **0.29**. This suggests that as the years of coding experience increase, there is a tendency for the yearly compensation in USD to increase as well, but the relationship is not very strong.
- For the **EUR (European Euro)**, the correlation is slightly stronger at **0.33**, but it is still a weak positive correlation.
- The **INR (Indian Rupee)** shows a moderate positive correlation of **0.47**. This suggests a more noticeable trend of increasing yearly compensation with increasing years of coding experience.
- The **GBP (Pound Sterling)** has the weakest positive correlation among the four currencies at **0.27**.

These findings suggest that while there is a general trend of increasing compensation with increasing years of coding experience, the strength of this relationship varies depending on the currency. This could be due to various factors such as differences in economic conditions, cost of living, and compensation practices in different countries.
untries.

It splits the values in the specified column on the ';' character, explodes the resulting lists into separate rows, and resets the index. It then counts the occurrence of each value (using `value_counts()`) and resets the index again to form a new DataFrame. The columns of the new DataFrame are named as the input column name and 'Count'.

2. **Applying the Function**: We applied the `clean_and_group` function to each of the three columns in our DataFrame, resulting in three grouped DataFrames: `df_LearnCode_grouped`, `df_LearnCodeOnline_grouped`, and `df_LearnCodeCoursesCert_grouped`.

3. **Displaying the DataFrames**: We defined a function `display_side_by_side(*args)` that takes any number of DataFrames as input and displays them side by side in a Jupyter notebook. We used this function to display our three grouped DataFrames side by side.


```python
# Function to clean and group the dataframe
def clean_and_group(df, column_name):
    df_clean = df[column_name].str.split(';').explode().reset_index()
    df_grouped = df_clean[column_name].value_counts().reset_index()
    df_grouped.columns = [column_name, 'Count']
    return df_grouped

# Copy the necessary columns from the original dataframe
df_learning = df[['LearnCode', 'LearnCodeOnline', 'LearnCodeCoursesCert']].copy()

# Clean and group each column
df_LearnCode_grouped = clean_and_group(df_learning, 'LearnCode')
df_LearnCodeOnline_grouped = clean_and_group(df_learning, 'LearnCodeOnline')
df_LearnCodeCoursesCert_grouped = clean_and_group(df_learning, 'LearnCodeCoursesCert')

# Function to display dataframes side by side
def display_side_by_side(*args):
    html_str = ''
    for df in args:
        html_str += df.to_html()
    display_html(html_str.replace('table', 'table style="display:inline;margin-right:10px;"'), raw=True)

# Display all 3 dataframes side by side
display_side_by_side(df_LearnCode_grouped, df_LearnCodeOnline_grouped, df_LearnCodeCoursesCert_grouped)
```

## 13. Popularity of JavaScript
Is JavaScript still the most popular programming language? Do you think it will still be popular in the future?

In this section, we have streamlined the process of analyzing the programming languages that developers have worked with and want to work with. We have defined a function `process_language_data` that takes a DataFrame and a column name as inputs, processes the data, and returns a cleaned DataFrame. 

This function is used to process both 'LanguageHaveWorkedWith' and 'LanguageWantToWorkWith' columns. It performs the following steps:

1. **Data Extraction and Cleaning**: The function extracts the relevant data from the specified column, removes any missing values, and splits the languages that are separated by semicolons.

2. **Data Transformation**: The function then explodes the DataFrame to create a new row for each language for every respondent. 

3. **Data Aggregation**: The function counts the occurrence of each language and calculates the percentage of respondents who have worked with or want to work with each language.

After processing the data with the `process_language_data` function, we merge the two resulting DataFrames on the `'Language'` column. The final DataFrame `df_language` contains the count and percentage of respondents who have worked with and want to work with each language. This analysis provides us with valuable insights into the popularity and demand of different programming languages.

```python
def process_language_data(df, column_name):
    df_temp = df[column_name].copy().dropna().str.split(';').explode()
    df_temp = df_temp.value_counts().reset_index()
    df_temp.columns = ['Language', f'{column_name} count']
    df_temp[f'Percentage {column_name}'] = (df_temp[f'{column_name} count'] / df_temp[f'{column_name} count'].sum()) * 100
    return df_temp

df_LanguageHaveWorkedWith_clean = process_language_data(df, 'LanguageHaveWorkedWith')
df_LanguageWantToWorkWith_clean = process_language_data(df, 'LanguageWantToWorkWith')

df_language = pd.merge(df_LanguageHaveWorkedWith_clean, df_LanguageWantToWorkWith_clean, on='Language')
```

In this section, we further refined our analysis to focus on the top 15 programming languages. We calculated the total percentage of developers who have worked with and want to work with each language. 

We then identified the top 15 languages based on the total percentage. This was done by summing the percentages o`f 'Have Worked Wit`h' an`d 'Want To Work Wit`h' for each language, and selecting the 15 languages with the highest total.

With this refined data, we created a side-by-side bar chart to visually compare the popularity and demand of these top 15 languages. The chart provides a clear comparison between the languages developers are currently using and the languages they are interested in learning or using in the future.

This visualization offers valuable insights into current trends in programming languages, highlighting the languages that are not only popular now, but also expected to be in demand in the future.

```python
language_totals = df_language.groupby('Language')[['Percentage LanguageHaveWorkedWith', 'Percentage LanguageWantToWorkWith']].sum()

# Sort by total count (sum of both columns) and select the top 15 languages
top_15_languages = language_totals.sum(axis=1).nlargest(15).index

# Filter the dataframe to include only the top 15 languages
language_totals_top_15 = language_totals.loc[top_15_languages]

language_totals_top_15.plot(kind='bar', figsize=(10, 6))
plt.title('Comparison of Top 15 Programming Languages (Have Worked With vs. Want To Work With)')
plt.xlabel('Language')
plt.ylabel('Total Count')
plt.legend(['Have Worked With', 'Want To Work With'])

plt.show()
```
